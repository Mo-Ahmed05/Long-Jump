# Data Exploratory Notebook

This notebook explores what the dataset include.

# Import Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
df = pd.read_csv('../Data/cleaned_long_Jump.csv', low_memory=False)
df[['Date', 'DOB']] = df[['Date', 'DOB']].apply(pd.to_datetime)
```

# Number of Samples Over Years

```python
dateCounter = df.groupby('Year').size().reset_index(name='Count')

x = dateCounter['Year']
y = dateCounter['Count']

sns.set_theme(style='white')
plt.figure(figsize=(12,6))

plt.plot(x, y, marker='o')

plt.xlabel('Year')
plt.title(f'Total Number of Samples Over the Years', fontsize=15)

ax = plt.gca()

ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax.yaxis.set_major_locator(ticker.MultipleLocator(2000))

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Age Distribution

### Age Distribution for Sample Size

```python
ageCounter = df.groupby('Age').size().reset_index(name='Count')

x = ageCounter['Age']
y = ageCounter['Count']

sns.set_theme(style='white')
plt.figure(figsize=(12,6))

plt.bar(x, y)

plt.xlabel('Age')
plt.title(f'Age Distribution of Samples', fontsize=15)

ax = plt.gca()

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax.yaxis.set_major_locator(ticker.MultipleLocator(2000))

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### Age Distribution for Competitors in the Sample

```python
age_count = df.groupby('Age')['Competitor'].nunique()

x = age_count.index
y = age_count.values

sns.set_theme(style='white')
plt.figure(figsize=(12,6))

plt.bar(x, y)

plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.title('Athletes Count per Age', fontsize=15)

ax = plt.gca()

ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax.yaxis.set_major_locator(ticker.MultipleLocator(500))

# Remove top/right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
